```python
def score_to_numeric(x):
    if x=='failed':
        return 0
    if x=='successful':
        return 1
    if x=='canceled':
        return 2
    if x=='undefined':
        return 3
    if x=='live':
        return 4
    if x=='suspended':
        return 5
```

```python
df['State_Code'] = df['state'].apply(score_to_numeric)
```
